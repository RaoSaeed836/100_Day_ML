```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("/content/covid_toy.csv")
df
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=56)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer(transformers=[
('tf1',SimpleImputer(),['fever']), # Corrected column name from 'Fever' to 'fever'
('tf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']), # Corrected column name from 'Cough' to 'cough'
('tf3',OneHotEncoder(drop='first',dtype=np.int32),['gender','city']) # Corrected column names and dtype
],remainder='passthrough')
```

```python
transformer.fit_transform(X_train)
```

```python
transformer.transform(X_test)
```

```python
df=df[['who','embarked','class','embark_town','age','deck']]
df
```

```python
df=df.drop(columns=['deck'])
```

```python
x=df.iloc[:,1:]
y=df['who']
x
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=67)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
transformer=ColumnTransformer(transformers=[
    ('tf1',OrdinalEncoder(categories=[['First','Second','Third']]),['class']),
    ('tf2',SimpleImputer(),['age']),
    ('tf3',OneHotEncoder(dtype=np.int32,drop='first'),['embarked','embark_town'])
],remainder='passthrough')
```

```python
transformer.fit_transform(X_train)
```

```python
transformer.transform(X_test)
```
