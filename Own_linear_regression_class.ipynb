```python
class Meralr:
  def __init__(self):
   self.m=None
   self.b=None
  def fit(self,X_train,y_train):
    num=0
    dum=0

    for i in range(X_train.shape[0]):
          num=num + ((X_train.shape[i] - X_train.mean())*(y_train.shape[i]- y_train.mean()))
          dum=dum + ((X_train.shape[i] - X_train.mean())*(X_train.shape[i] - X_train.mean()))
          self.m=num/dum
          self.b=y_train.mean()-(self.m* X_train.mean())
  def predict(self,X_test):
           return self.m* X_test + self.b
```
